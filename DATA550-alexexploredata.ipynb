Destinations I struggled with the syntax. I think it makes sense to stick to only visualizing delays and such for the destinations that AQ services.

```python
#What is AQ's cancelled flights rate vs the other airlines?

flightCnt=air.groupby(['UniqueCarrier'])['Month'].count()
canFlightCnt=air[np.isnan(air.AirTime)].groupby(['UniqueCarrier'])['Month'].count()
divFlightCnt=air.groupby(['UniqueCarrier'])['Diverted'].sum()
combinedDF=pd.DataFrame()

combinedDF["FlightCnt"] = flightCnt
combinedDF["CanFlightCnt"] = canFlightCnt
combinedDF["DivertedCnt"] = divFlightCnt #Number of diverted Flights in 2018
combinedDF["CancelRate"] = combinedDF.CanFlightCnt / combinedDF.FlightCnt
combinedDF["DivertedRate"] = combinedDF.DivertedCnt / combinedDF.FlightCnt
print(combinedDF.sort_values("CancelRate"))
```
